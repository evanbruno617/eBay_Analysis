```python
import pandas as pd
```

```python
stick_cpi = pd.read_csv('Economic_data/CORESTICKM159SFRBATL.csv')
stick_cpi = stick_cpi.rename(columns={'CORESTICKM159SFRBATL': 'Sticky_Price_CPI'})
stick_cpi['Change_Sticky_Price_CPI'] = (stick_cpi['Sticky_Price_CPI'] - stick_cpi['Sticky_Price_CPI'].shift(1)) / stick_cpi['Sticky_Price_CPI']
stick_cpi
```

```python
ms = pd.read_csv('Economic_data/M2REAL.csv')
ms = ms.rename(columns={'M2REAL': 'Money_Supply'})
ms['Change_Money_Supply'] = (ms['Money_Supply'] - ms['Money_Supply'].shift(1)) / ms['Money_Supply']
ms
```

```python
med_cpi = pd.read_csv('Economic_data/MEDCPIM158SFRBCLE.csv')
med_cpi = med_cpi.rename(columns={'MEDCPIM158SFRBCLE': 'Median_CPI'})
med_cpi['Change_Median_CPI'] = (med_cpi['Median_CPI'] - med_cpi['Median_CPI'].shift(1)) / med_cpi['Median_CPI']
med_cpi
```

```python
pce = pd.read_csv('Economic_data/PCE.csv')
pce = pce.rename(columns={'PCE': 'Personal_Consumption_Expenditure'})
pce['Change_Personal_Consumption_Expenditure'] = (pce['Personal_Consumption_Expenditure'] - pce['Personal_Consumption_Expenditure'].shift(1)) / pce['Personal_Consumption_Expenditure']
pce
```

```python
rpce = pd.read_csv('Economic_data/PCEC96.csv')
rpce = rpce.rename(columns={'PCEC96': 'Real_Personal_Consumption_Expenditure'})
rpce['Change_Real_Personal_Consumption_Expenditure'] = (rpce['Real_Personal_Consumption_Expenditure'] - rpce['Real_Personal_Consumption_Expenditure'].shift(1)) / rpce['Real_Personal_Consumption_Expenditure']

rpce
```

```python
int_r = pd.read_csv('Economic_data/REAINTRATREARAT10Y.csv')
int_r = int_r.rename(columns={'REAINTRATREARAT10Y': '10_year_interest'})
int_r['Change_10_year_interest'] = (int_r['10_year_interest'] - int_r['10_year_interest'].shift(1)) / int_r['10_year_interest']
int_r
```

```python
unrate = pd.read_csv('Economic_data/UNRATE-2.csv')
unrate = unrate.rename(columns={'UNRATE': 'Unemployment_Rate'})
unrate['Change_Unemployment_Rate'] = (unrate['Unemployment_Rate'] - unrate['Unemployment_Rate'].shift(1)) / unrate['Unemployment_Rate']
unrate
```

```python
total_df = pd.merge(unrate, int_r, on='DATE', how='outer')
total_df = pd.merge(total_df, rpce, on='DATE', how='outer')
total_df = pd.merge(total_df, med_cpi, on='DATE', how='outer')
total_df = pd.merge(total_df, pce, on='DATE', how='outer')
total_df = pd.merge(total_df, ms, on='DATE', how='outer')
total_df = pd.merge(total_df, stick_cpi, on='DATE', how='outer')
total_df = total_df.rename(columns={'DATE': 'Date'})
total_df = total_df.loc[total_df['Date'] > '2021-01-01']
total_df['Date'] = pd.to_datetime(total_df['Date'])
total_df.dropna(inplace=True)
total_df
```

```python
total_df.to_csv('Economic_Data.csv', index=False)
```
